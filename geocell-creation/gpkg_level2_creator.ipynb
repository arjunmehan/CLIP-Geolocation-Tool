```python
import geopandas as gpd
from fiona import listlayers
```

```python
gpkg_file = "../resources/gadm_410.gpkg"
layers = listlayers(gpkg_file)
print(layers)
```

```python
gdf = gpd.read_file(gpkg_file, layer='gadm_410')
print(gdf.head())
```

### This project only requires the polygons of the level 2 subdivisions

```python
level2 = gdf[['GID_0', 'NAME_0', 'GID_1', 'NAME_1', 'GID_2', 'NAME_2', 'geometry']].copy()
level2 = level2.dissolve(by=['GID_2', 'NAME_2', 'GID_1', 'NAME_1', 'GID_0', 'NAME_0']).reset_index()
print(level2.head())
print("done!")
```

```python
level2.to_file("../resources/gadm_level2.gpkg", driver="GPKG")
```
