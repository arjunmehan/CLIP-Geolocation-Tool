```python
import pandas as pd
import geopandas as gpd
```

```python
df_images = pd.read_csv('../resources/mp16_combined_clean.csv')
```

```python
gdf = gpd.read_file('../resources/gadm_split_2.gpkg')
```

```python
gdf
```

```python
pts = gpd.GeoDataFrame(df_images, geometry=gpd.points_from_xy(df_images['LON'], df_images['LAT']), crs="EPSG:4326")
```

```python
joined = gpd.sjoin(pts, gdf[['geometry']], how='left', predicate='within')
```

```python
df_images['geocell_index'] = joined['index_right'].values
```

```python
df_images
```

```python
df_images.to_csv('../resources/mp16_w_index.csv', index=False)
```
